# Books Information
This notebook reads a csv file containing a list of ISBN numbers, queries the Google Books site and scrapes data about the book

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
book_isbn_df = pd.read_csv('Resources/ISBNList2.csv',dtype=str)
```

```python
book_isbn_df['Title'] = ''
book_isbn_df['Year'] = ''
book_isbn_df['Author'] = ''
book_isbn_df['Pages'] = ''
book_isbn_df['Synopsis'] = ''
book_isbn_df['Common terms and phrases'] = ''
book_isbn_df['Subject'] = ''
book_isbn_df.head()
```

```python
for index in range(len(book_isbn_df['ISBN'])):
    try:
        ISBN = book_isbn_df.iloc[index]['ISBN']
        URL = f'https://www.google.com/search?q={ISBN}&tbm=bks&ei=JoNFXvmZMKupytMPxuCTCA&ved=0ahUKEwj5xNrjgs_nAhWrlHIEHUbwBAEQsJ4FCAM'
        result  = requests.get(URL)
        src= result.text
        soup = BeautifulSoup(src,'lxml')
        elements = soup.find_all("div",{'class':'kCrYT'})
        #print (src)
        URL_Before_redirect = elements[1].find('a')['href'] + '&newbks=1&newbks_redir=1'
        print(URL_Before_redirect)
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:54.0) Gecko/20100101 Firefox/54.0'}
        new_result = requests.get(URL_Before_redirect,headers=headers,allow_redirects=True)
        # print(new_result.text)
        src= new_result.text
        new_soup = BeautifulSoup(src,'lxml')
        title = new_soup.find('div',{'class':'zNLTKd'}).text
        print(title)
        book_isbn_df.loc[index]['Title'] = title
        #year
        year = new_soup.findAll('span',{'class':'LrzXr kno-fv'})[2].text
        print(year)
        book_isbn_df.loc[index]['Year'] = year
        #author
        try:
            author = new_soup.findAll('div',{'class':'WJQQof'})[3].text
            book_isbn_df.loc[index]['Author'] = author
        except:
            pass

        #pages
        pages = new_soup.findAll('span',{'class':'LrzXr kno-fv'})[1].text
        print(pages)
        book_isbn_df.loc[index]['Pages'] = pages
        #synopsis
        try:
            synopsis = new_soup.find('div',{'class':'FhLkgb'}).text
            print(synopsis)
            book_isbn_df.loc[index]['Synopsis'] = synopsis
        except:
            pass
        #Common terms and phrases
        common_terms = new_soup.findAll('entity-page-viewport-entry')
        append_common_terms = ''
        for term in common_terms:
            try:
                append_common_terms = append_common_terms + str(term['data-search-query']) + ', '
            except:
                pass
        print(append_common_terms)
        book_isbn_df.loc[index]['Common terms and phrases'] = append_common_terms
        #subject
        try:
            subject = new_soup.findAll('span',{'class':'LrzXr kno-fv'})[7].text
            print(subject)
            book_isbn_df.loc[index]['Subject'] = subject
        except:
            pass
    except:
        pass
```
