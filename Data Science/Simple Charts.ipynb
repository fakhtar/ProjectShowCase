```python
# imports
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import random
```

```python
#Sample tempratures in farenhites
farenhite_temps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
# populate x axis list
x_axis = np.arange(1, 13, 1)
# convert farenhite temrature to celsius via lambda function
celsius_temp = [(lambda F: (F-32) * 0.56)(x) for x in farenhite_temps]
```

# Line Chart

```python
fhite_plot, = plt.plot(x_axis,farenhite_temps,marker ='o', color='blue', label='Farenhite')
cels_plot, = plt.plot(x_axis,celsius_temp,marker ='^', color='green', label='Celcius')
#generate legent
plt.legend()
plt.title("Farenhite and Celsius Chart Comparison")
plt.xlabel('Month')
plt.ylabel('Degrees')
plt.grid()
```

# Bar Chart

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
plt.bar(cities,bars_in_cities)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, 10)
plt.title('Number of bars in each city')
plt.xlabel('City names')
plt.ylabel('Number of bars')
```

# Pie Chart

```python
my_data = {'labels':['Shirts','Pants','Trousers','Flip-Flops','Jackets','Toys'],'colors':['red','yellow','green','blue','orange','purple'],'values':[25,52,12,22,11,5],'explode':[.1,0,0,0,.2,.3]}
plt.pie(
    x=my_data['values'],
    explode=my_data['explode'],
    labels=my_data['labels'],
    colors=my_data['colors'],
    autopct=None,
    pctdistance=0.6,
    shadow=True
)
plt.title('Breakdown of Sales by Item')
```

# Scatter Plot

```python
x_axis = np.arange(0,100,1)
values = [random.random() for x in x_axis]
my_c = [random.random()*x for x in x_axis]
plt.scatter(
    x_axis,
    values,
    s=x_axis,
    c=my_c,
    marker=None,
    cmap='PuBuGn',
    norm=None,
    vmin=None,
    vmax=None,
    alpha=None,
    linewidths=None,
    verts=None,
    edgecolors=None,
    plotnonfinite=False
)
plt.title('Simple Scatt')
```
